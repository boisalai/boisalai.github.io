## Training modules overview
http://docs.fast.ai/training.html

```python
# Walk-through of key functionality
from fastai.docs import *
mnist_data = get_mnist()
```

```python
# Basic training with Learner
model = simple_cnn((3,16,16,2))
learn = Learner(mnist_data, model)
learn.fit(1)
```

```python
# Viewing metrics
learn.metrics=[accuracy]
learn.fit(1)
```

```python
# Extending training with callbacks
cb = OneCycleScheduler(learn, lr_max=0.01)
learn.fit(1, callbacks=cb)
```

```python
learn.recorder.plot_lr(show_moms=True)
```

```python
# Extending Learner with train
learn.fit_one_cycle(1)
```

```python
# Applications
from fastai.vision import *
learn = ConvLearner(mnist_data, tvm.resnet18, metrics=accuracy)
learn.fit_one_cycle(1)
```
